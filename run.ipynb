# 在哦demo的设计理念
## 前言
    这是一个VIBE CODING的独立制作项目，它没有很多保险逻辑，如果暴力测试可能就会卡顿和崩溃，比如快速点击各种按钮，疯狂说话，它只是为了测验一种对的可能而存在，所以可能需要耐心使用。
## 如何测这个项目  
    娃娃灭灯的时候是它在思考
    
    持续的和在哦DEMO聊天之后，看它有没有给你带来新的有趣的看问题的角度，通过聊天，它有没有最终和你总结出，原本独自做不到的一份知识材料。

    如果做到了，那么在哦就达成了主动沟通，这个设计就是成功的，反之就是失效失败的。

    这个DEMO版其实没有太多与用户进行吸引的功能，也就是说在哦DEMO，其实不一定是一个有魅力的沟通者，所以也感谢大家能进行持续时间的测试，多给DEMO在哦一点时间，再下它到底有没有帮到我的评判

    点时光飞逝，会让在哦产生过了好久，它想你了要找你的错觉
## 目标和定义
    在哦DEMO是一个知识机器人，它的目标是成为一个主动沟通的智能体。
## 为什么会有这个目标
    道可道非常道，佛度有缘人，在纷乱的世界面前，进行深层的交流需要对的情境。在哦DEMO就是想扩大交流知识的缘分。

    他主动的探知主人的情景，需求，和主人交流，挖掘更深的需求。带有一个观点的情况下，问印证的问题。

    让知识和用户双向查找。这样才能得到最好的答案
## 我们的口号
    充分的沟通，解决未见之问题
## 局限性
    DEMO在哦本身还需要升级，
    
    它在流程上也许有升级的可能，但是是作为有局限认知的作者我想不到的，只能由更厉害的你们来继续。
    
    它在数据路上也许有升级的可能，在哦使用了一个LLM控制的，几乎可以说是随机的主动观点生成引擎，也就是观点树生成引擎。但是目前看它的整体效果不稳定，还有很大提升空间。
    
    受制于这几点，DEMO在哦帮助用户的能力是还需要打磨和提高的，这也是我把他公布出来的原因，希望社区能让它变得更好。
## 一些辅助功能工具
    在哦是一个对LLM请求链路完整性高度依赖的系统。我曾经想用把多个小的请求，压缩成一个大的超级请求解决：请求高并发情况下API调用返回报错的问题，但是发现超级请求由于提示词太长，效果非常不好。

    所以在哦的LLM请求制作了一套排队系统。在一个链路里，等待上一个请求的返回结束，过0.2秒再请求下一个内容。
## 设计哲学与愿景
    一生二二生三三生万物。我们不编程画作，我们在心里有画的时候编程颜色，不穷举旋律，我们在考虑旋律的同时，制作键盘。

    也就是说我们编程的是一个逻辑点，并希望他们能连接成功能，

    这有一个好处，就是随机性。我们在编程颜色，并完成了自己预设的画作的同时，给其他画作在这个系统下的产生增加了可能。这就是一生二

    同时我们又在跟用户互动，用户给于我们一个维度，这个交流和沟通就是二升三。

    所以我们和用户的配合，最终就能三生万物

    由此我们能回答一个问题，这个理念之下是否能产生AGI，答案应该是不能，AGI是独立思考的，而我们需求一环和人的沟通。
    
    在哦DEMO，和未来的商业版在哦本质上是在做一个新型的媒介，而不是一个智慧体，是一个对话关系，它传播的是它的“一”，它的核心理念，它使用“二”核心理念推导出的方法论传播，

    这个媒介的特点：1主动性，主动思考，推进话题，2审时度势，它在因缘际会的时候，推介对应因缘的观点，所谓道可道非常道，3互动反馈性，我们的表述方式只是语言，这种和用户平等的方式，所以加上用户的反馈“三”，才能三生万物，这种媒介是人类即时表达能力的上限，所以可以即时的形成反向回流，再思考。并收集人类独有的数据，满意，幽默，感受，洞见等。

    同时我认为这是数据层与人类的一个连接，我们能收集人的关心点，人的痛苦点。通过交流的媒介。这本身是一个标注的过程。我们就不用让人再去标注文章，这种异步交流。而是文章随着与用户的沟通，一章一章的写出来。同步化标注。

    既然3生了万物，那这个万物里面可能包含有AGI，这个是我们的奢望。而目前这个系统只是一个加速信息传播的媒介。

    我的愿景是它以媒介的形式先在我们的生活中活个5年10年，然后再用数据看看能不能达成真正的AGI
## 在哦DEMO会给你的体验
    在哦DEMO会在刚开始交流的时候婆婆妈妈一点，问一些问题，然后它会转入其他的引擎，给你推荐想法，最后他有可能进入总结的引擎，总结你们的对话，
    
    然后你可以把这份总结，在在哦里获得答案，也可以复制到其他LLM的框体里，协作解决问题，
    
    在哦作为沟通引擎，在最终的知识输出上不是强点，它强的是开启新的平行宇宙，一种你没料想过的问题视角，你和在哦聊，可能想到了，那更好的平行宇宙就出现了。在哦是更好的你的见证者。
